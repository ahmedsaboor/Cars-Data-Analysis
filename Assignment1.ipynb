```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.transforms as mtransforms
import matplotlib.lines as mlines
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import r2_score
import seaborn as sb
%matplotlib inline
```

```
df = pd.read_csv("cars.csv")
df = df.drop('Unnamed: 0', 1)
df.head()
```

```
fig, ax = plt.subplots(figsize = (16,11))
sb.scatterplot(x = "year", y = "price", hue = "title_status", data = df, ax = ax)
ax.set_xlabel("Vehicle Registration Year")
ax.set_ylabel("Sale Price($)")
plt.show()
```

```
fig, (ax0, ax1) = plt.subplots(nrows=1,ncols=2, figsize=(16,9))
#sb.set_style("darkgrid")
sb.scatterplot(x = "price", y = "mileage", data = df, ax =ax0)
sb.scatterplot(x = "year", y = "mileage", data = df, ax= ax1)
ax0.set_xlabel("Sale Price($)")
ax1.set_xlabel("Vehicle Registration Year")
ax0.set_ylabel("Distance Travelled (Miles)")
ax1.set_ylabel("Distance Travelled (Miles)")
plt.show()
```

```
col_list = ['price']
col_list1 = ['mileage']
col_list2 = ['brand']
col_list3 = ['year']
val1 = pd.read_csv("cars.csv", usecols=col_list)
val2= pd.read_csv("cars.csv", usecols=col_list1)
val3= pd.read_csv("cars.csv", usecols=col_list2)
val4= pd.read_csv("cars.csv", usecols=col_list3)
```

```
fig = plt.figure()
ax = fig.add_axes([0,0,5,5])
yx = val1['price']
xx = val3['brand']
ax.bar(xx,yx)
plt.title('Price vs. Brand', fontsize=50)
plt.xlabel('Brand', fontsize=30)
plt.ylabel('Price', fontsize=30)
ax.tick_params(axis='both', which='major', labelsize=16)
ax.tick_params(axis='both', which='minor', labelsize=15)
plt.show()
```

```
fig = plt.figure()
ax = fig.add_axes([0,0,2,2])
y1 = val1['price']
x1 = val4['year']
ax.bar(x1,y1)
plt.title('Price vs Year')
plt.xlabel('Price')
plt.ylabel('Year')
plt.show()
```
